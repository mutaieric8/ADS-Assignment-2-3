```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_attrition=pd.read_csv('https://raw.githubusercontent.com/Africa-Data-School/ADS-Assignment-2-3/main/WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

Assignment 1. Show a breakdown of distance from home by job role and attrition.

```python
df_distance_bd=df_attrition.groupby(['JobRole','Attrition'])['DistanceFromHome'].sum().reset_index()
df_distance_bd
```

Assignment 2. Compare average monthly income by education and attrition

```python
df_davg_income=df_attrition.groupby(['EducationField','Attrition'])['MonthlyIncome'].mean().reset_index()
df_davg_income
```

Assignment 3.
Based on your analysis in Assignment 2, express your results in the form of the most appropriate 
visualization using any visualization library of your choice.

```python
fig=plt.figure(figsize=(12,12))
sns.barplot(data=df_distance_bd,x='Attrition',y='DistanceFromHome',hue='JobRole')
plt.title("Attrition in relation to Distance from Home and Job Role")
plt.xlabel("Attrition")
plt.ylabel("Distance from Home")
```

```python
fig=plt.figure(figsize=(12,12))
sns.barplot(data=df_davg_income,x='EducationField',y='MonthlyIncome',hue='Attrition')
plt.title("Attrition in relation to Education and Monthly Income")
plt.xlabel("Education")
plt.ylabel("Monthly Income")
```
